```python

import RELUNTK
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()
import pandas as pd 
import torch.optim
import torch
%load_ext autoreload
%autoreload 2
```

```python
train_data = pd.read_csv("Simple_train.csv" )

Xtrain = torch.tensor(train_data["X"].values.reshape(-1,1) ,dtype = torch.float)
Ytrain = torch.tensor(train_data["Y"].values.reshape(-1,1) , dtype = torch.float)


plt.scatter(Xtrain, Ytrain )
```

```python
nhidden = 50
model = RELUNTK.NTKtest(1,50,1)
lr = 0.01
optimizer = torch.optim.SGD(model.parameters(), lr = lr)
epochs = 100 
for i in range(epochs):
    loss = RELUNTK.train(model , Xtrain , optimizer , Ytrain , nhidden)
    
    if i % 10==0 or i == epochs-1:
        print("Epoch %5d/%5d , loss = %5.3f" , i+1 , epochs , loss)
    

```
